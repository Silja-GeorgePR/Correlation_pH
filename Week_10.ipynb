### 4. Carry out correlation analysis on the dataset and find the correlation factor between
a. ph and Chloramines 

b. ph and Conductivity

```python
corrmatrix=data.corr()
plt.subplots(figsize=(20,15))
sns.heatmap(corrmatrix, vmin=-.7,vmax=.7,linewidths=.2,annot=True,cmap='YlGnBu')
```

No two features in the data set are highly correlated

```python
data[['ph','Chloramines']].corr()
```

```python
sns.scatterplot(data=data,x=data['ph'], y=data['Chloramines'])
```

The correlation factor is -0.02468,approximated to 0 itself,thus we can conclude that there is no significant corrrelation between ph and chloramines

```python
data[['ph','Conductivity']].corr()
```

```python
sns.scatterplot(data=data,x=data['ph'], y=data['Conductivity'])
```

The correlation factor is 0.014128, which is also approximated to 0, so we can conclude that there is no significant correlation between ph and conductivity
